#A Bayesian Approach: What is the probability that a male patient takes longer to fund than a female patient?

A Bayesian approach allows us to make direct probability statements. The interpretation of its results is simpler than interpreting p-values or confidence intervals. 

The main drawback to a Bayesian approach is that we must specify our prior belief about our unknown parameter, and this inserts some subjectiveness into our scientific process. Bayesians respond that we can create very weak priors that don't much affect the results, priors that will allow the data to dominate the final result.

In this case, our priors are a) the number of hours to fund an average male patient, and b) the number of hours to fund an average female patient. We will test various priors to see if they affect our final result much. Looking at the data to inform our 'prior' belief is a method called Empirical Bayes.

The data we have is the amount of time it takes to fund a profile. The random variable I think that's most similar to is the Exponential. The story for the Exponential, as taken from [Joe Blitzstein's Intro to Probability course](http://www.stat110.net), is this:

> You are waiting in line at the supermarket. The time it takes for each person to finish their checkout can be modeled as $Exponential(\lambda)$, where $1/\lambda$ is the average amount of time for someone to finish their checkout.

We can think of our data, the number of hours to fund a profile, as coming out of a random number generator. This number generator is given the expected number of hours for an average profile to be funded, and it gives us sample figures. In this sense, we view the data as a random variable. (In the classical context, the data is not random. It either is a certain figure, or it isn't.)

Our main violation of the Exponential distribution is that each profile is not equally likely to be funded in the same amount of time. Profiles with lower funding request amounts are funded more quickly, although the correlation is not that strong.

##Conjugate prior

If our data is in $Exponential(\lambda)$ form, if we use the conjugate prior distribution to represent our prior belief, the math will be clean. The conjugate prior here is a $Gamma(\alpha,\lambda)$ distribution, where the $\lambda$s in both distributions are identical.

The inutitive story for the Gamma distribution is

> We are again in line at the supermarket. There are $\alpha$ people in front of us, in a single line. Each amount of time for each person to finish their checkout can be modeled by $Exponential(\lambda)$, where $1/\lambda$ is the average amount of time to finish a checkout. The probability that everyone will finish their checkouts by $X$ total time is modeled by $Gamma(\alpha, \lambda)$.

Let's model the time to fund a profile via the $Gamma(\alpha, \lambda)$ distribution. If we just want to see how long it will take for one profile to be funded, that is the same as the $Exponential(\lambda)$ distribution.

```python
def plot_gamma(num_profiles, hrs_to_fund, a=0, b=100):
	"""num_profiles is # of occurrences. hrs_to_fund is E[X] for Expo(lambda). a and b are x-axis limits."""
	x = np.linspace(a, b, 10000)
	lambda_ = 1./hrs_to_fund # how many profiles funded per hour; rate of lambda_ profiles funded per hour; next profile funded in 1/lambda_ hours
	y = stats.gamma.pdf(x, a=num_profiles, scale=1./lambda_)

	plt.plot(x, y)
	plt.fill_between(x, 0, y, alpha=0.4)
	plt.xlabel('hours until all {} profiles funded when we expect {:.2f} profiles funded every hour, {:.2f} hours until the next profile funded'.format(num_profiles, lambda_, 1./lambda_))

	print stats.gamma.cdf(1, a=num_profiles, scale=1./lambda_), 'probability after 1 hr'
	print stats.gamma.cdf(num_profiles*hrs_to_fund, a=num_profiles, scale=1./lambda_), 'probability after {} hrs'.format(hrs_to_fund*num_profiles)
```

```python
plot_gamma(num_profiles=1, hrs_to_fund=df.hrs_to_fund.mean())
```

```python
x = np.linspace(0, 100, 10000)
lambda_ = lambda_ # how many occurences per unit time; lambda_ per minute; next occurence in 1/lambda_ time
y = stats.expon.pdf(x, scale=1./lambda_)

plt.plot(x, y)
plt.fill_between(x, 0, y, alpha=0.4)
plt.xlabel('hours until the next occurence when we expect {} occurrences every hour, {:.2f} hours until the next occurrence'.format(lambda_, 1./lambda_))

print stats.expon.cdf(1, scale=1./lambda_), 'probability after 1 hr'
print stats.expon.cdf(hrs_to_fund, scale=1./lambda_), 'probability after {} hr'.format(hrs_to_fund)
```

####Running the model

```python
p = pm.Gamma('p', 1, 1./lambda_)

occurrences = df[df.gender == 'male'].hrs_to_fund

obs = pm.Exponential("obs", p, value=occurrences, observed=True)

# To be explained in chapter 3
mcmc = pm.MCMC([p, obs])
mcmc.sample(18000, 1000)
```

```python
with sns.axes_style('darkgrid'):
    #figsize(12.5, 4)
    plt.title("Posterior distribution of $p_{male}$, the true num of hrs to fund all male profiles")
    plt.hist(mcmc.trace("p")[:]**-1, bins=25, histtype="stepfilled", normed=True)
    plt.legend()
```

```python
lambda_, df.hrs_to_fund[df.gender == 'female'].median(), df.hrs_to_fund[df.gender == 'male'].median()
```

```python
# prior for mean male and mean female. i guess if i'm setting mu to the mean of both, that's empirical bayes

lambda_male = 1/df.hrs_to_fund[df.gender == 'male'].median()
lambda_female = 1/df.hrs_to_fund[df.gender == 'female'].median()

lambda_male = 1/df.hrs_to_fund[df.gender == 'male'].mean()
lambda_female = 1/df.hrs_to_fund[df.gender == 'female'].mean()

#p_male = pm.Normal('p_male', mu=80, tau=1./variance)
#p_female = pm.Normal('p_female', mu=80, tau=1./variance)

#mu = df.hrs_to_fund.mean()
#p_male = pm.Normal('p_male', mu=mu, tau=1./50)
#p_female = pm.Normal('p_female', mu=mu, tau=1./50)

p_male = pm.Gamma('p_male', 1, lambda_)
p_female = pm.Gamma('p_female', 1, lambda_)

@pm.deterministic
def delta(p_male=p_male, p_female=p_female):
    return p_male**-1 - p_female**-1

obs_male = pm.Exponential("obs_male", p_male, value=df.hrs_to_fund[df.gender == 'male'], observed=True)
obs_female = pm.Exponential("obs_female", p_female, value=df.hrs_to_fund[df.gender == 'female'], observed=True)

mcmc = pm.MCMC([p_male, p_female, delta, obs_male, obs_female])
mcmc.sample(20000, 1000)

p_male_samples = mcmc.trace("p_male")[:]**-1
p_female_samples = mcmc.trace("p_female")[:]**-1
delta_samples = mcmc.trace("delta")[:]
```

```python
with sns.plotting_context("paper", font_scale=2.5):
    plt.figure(figsize=(20,10))
    ax = plt.subplot(211)

    plt.hist(p_male_samples, histtype='stepfilled', bins=25, alpha=0.85,
             label="posterior of $1/\lambda_{male}$", color="#A60628", normed=True)
    plt.legend(loc="upper right")
    plt.title("Posterior distributions of $1/\lambda_{male}$, $1/\lambda_{female}$, and $1/\lambda_{male} - 1/\lambda_{female}$")
    
    plt.hist(p_female_samples, histtype='stepfilled', bins=25, alpha=0.85,
             label="posterior of $1/\lambda_{female}$", color="#467821", normed=True)
    plt.legend(loc="upper right")
    plt.xlabel('Average number of hours to fund a profile, $1/\lambda$')
    plt.tight_layout()

    ax = plt.subplot(212)
    plt.xlim(0, 40)
    plt.hist(delta_samples, histtype='stepfilled', bins=25, alpha=0.85,
             label="posterior of $1/\lambda_{male} - 1/\lambda_{female}$", color="#7A68A6", normed=True)
    plt.legend(loc="upper right")
    plt.xlabel('Difference between the average number of hours to fund a profile')
```
