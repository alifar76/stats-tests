#A Bayesian Approach: What is the probability that it takes longer to fund a male profile than a female profile?

A Bayesian approach allows us to make direct probability statements. The interpretation of its results is simpler than interpreting p-values or confidence intervals. 

The main drawback to a Bayesian approach is that we must specify our prior belief about the situation, and this inserts some subjectiveness into our calculation. Bayesians respond that we can create very weak priors that don't much affect the results, which will allow the data to dominate the final result.

In this case, our priors are a) the average number of hours to fund a male patient, and b) the average number of hours to fund a female patient. Because we looked at the data to inform our 'prior' belief, this method is called Empirical Bayes. We will test various priors to see if they affect our final result much.

The data we have is the amount of time it takes to fund a profile. For each profile, we can think of the time it takes to be funded as an Exponential random variable. The story for the Exponential, as taken from [the notes for](https://bayesrule.files.wordpress.com/2014/07/probability_cheatsheet_140718.pdf) Joe Blitzstein's [Intro to Probability course](http://www.stat110.net), is:

> You are in a grassy field, looking at shooting stars. The time in between each shooting star can be modeled as $Exponential(\lambda)$, where $1/\lambda$ is the expected amount of time until a shooting star passes.

We can think of our data, the number of hours to fund a profile, as coming out of a random number generator. This number generator is given the average number of hours for a profile to be funded, and it gives us sample numbers. In this sense, we view the data as a random variable. (In the classical context, the data is not random. It either is a certain figure, or it isn't.)

##Assumption violations of the Exponential distribution

In the example, we are measuring the time between each shooting star. But when we measure time-to-fund, it's not that one profile is posted onto the website, then after it's funded, another profile is posted, and we wait for that to be funded. Instead, there are multiple profiles posted to the site, and we are starting our timer from when the profile is posted to when it is funded. In the lingo, we are not strictly looking at inter-arrival times.

An important property of the Exponential distribution is that it is memoryless. The time it took for one shooting star doesn't affect the time until the next shooting star. But for us, a donor might fully fund multiple profiles in a session, causing the latter profiles' time-to-fund to be affected by its predecessors'. This violation of the memoryless property stems from the fact that we're not strictly measuring interarrival times.

Another assumption we violate is that the times-to-fund are identically distributed. To us sitting in a grassy field, all the shooting stars are the same. A certain kind of shooting star does not occur more frequently than another. But our profiles can more easily be categorized into various groups. For example, profiles that request more money take longer to fund.

##Conjugate prior; Empirical Bayes

If our data fits an Exponential distribution, what should our distribution be to state our prior belief of the average number of hours to fund a male profile and a female profile? If we use a conjugate prior, than our final result will be analytically tractable, and we will be able to come to a mathematically clean solution.

The conjugate prior here is a $Gamma(\alpha,\lambda)$ distribution, where $\lambda$ is the same as in the $Exponential(\lambda)$ distribution.

The inutitive story for the Gamma distribution is:

> We are again looking at shooting stars. We want to wait to see $\alpha$ stars before we go home. The amount of time between each shooting star can be modeled as $Exponential(\lambda)$, where $1/\lambda$ is the expected amount of time until a shooting star passes. The probability that all $\alpha$ shooting stars will occur by $X$ total time is modeled by $Gamma(\alpha, \lambda)$.

Let's model the time to fund a single profile via the $Gamma(\alpha, \lambda)$ distribution, where $\alpha=1$. If we want to see how long it will take for one profile to be funded, that is the same as the $Exponential(\lambda)$ distribution. $Gamma(1, \lambda) = Exponential(\lambda)$

```python
def plot_gamma(num_profiles, hrs_to_fund, a=0, b=100):
	"""num_profiles is # of occurrences. hrs_to_fund is E[X] for Expo(lambda). a and b are x-axis limits.
    hrs_to_fund = 1/lambda_!!!"""
	x = np.linspace(a, b, 10000)
	lambda_ = 1./hrs_to_fund # how many profiles funded per hour; rate of lambda_ profiles funded per hour; next profile funded in 1/lambda_ hours
	y = stats.gamma.pdf(x, a=num_profiles, scale=1./lambda_)

	plt.plot(x, y)
	plt.fill_between(x, 0, y, alpha=0.4)
	plt.xlabel('hours until all {} profiles funded when we expect {:.2f} profiles funded every hour, {:.2f} hours until the next profile funded'.format(num_profiles, lambda_, 1./lambda_))
	plt.ylabel('Probability')
    
	print stats.gamma.cdf(1, a=num_profiles, scale=1./lambda_), 'probability after 1 hr'
	print stats.gamma.cdf(num_profiles*hrs_to_fund, a=num_profiles, scale=1./lambda_), 'probability after {} hrs'.format(hrs_to_fund*num_profiles)
```

```python
plot_gamma(num_profiles=1, hrs_to_fund=df.hrs_to_fund.mean())
```

```python
hrs_to_fund = df.hrs_to_fund.mean()
lambda_ = 1./hrs_to_fund

x = np.linspace(0, 100, 10000)
y = stats.expon.pdf(x, scale=1./lambda_)

plt.plot(x, y)
plt.fill_between(x, 0, y, alpha=0.4)
plt.xlabel('hours until the next occurence when we expect {} occurrences every hour, {:.2f} hours until the next occurrence'.format(lambda_, 1./lambda_))
plt.ylabel('Probability')

print stats.expon.cdf(1, scale=1./lambda_), 'probability after 1 hr'
print stats.expon.cdf(hrs_to_fund, scale=1./lambda_), 'probability after {} hr'.format(hrs_to_fund)
```

##Using the conjugate prior Gamma() to get the posterior distribution Gamma(); Empirical Bayes

```python
plot_gamma(1, df.hrs_to_fund.median())
```

The above, $Gamma(1, 46.36)$, is my prior. I have information on only 1 profile, and I think it will take 46 hrs to fund it, which is the median number of hours to fund all profiles, male and female. I chose the median instead of the mean for the expected value of how many hours until a profile is funded because I think it's less affected by extreme values and high variance that this distribution has.

I chose 1 profile arbitrarily. If $\alpha=1$, the prior is not very strong, which is what I want. I could also give more weight to my prior, by increasing the number of profile I've 'seen' before the experiement. This is where subjective priors come into play.

I can check later how much of a difference it makes to give more weight to my prior.

####Now, w the data I have

Our posterior distribution, because I used the $Gamma(\alpha, \lambda)$ conjugate prior, is $Gamma(\alpha + n, \lambda + n\bar x)$ for one $Exponential(\lambda)$ experiment, or one new profile that I've learned about.

So, let's say that we have a new $Exponential(\lambda)$ datapoint, and it took 75 hours to fund it.

My new posterior distribution is $Gamma(1 + 1, \frac{46 + 75}{2})$.

$46 + 2 * \frac{46 + 75}{2}$

$\frac{46 * 1}{n + 1} + \frac{75 * n}{n + 1}$

It seems to make more sense to have the new `hrs_to_fund` be an average of the two figures, rather than just a sum. A sum will blow up when `n` gets large, and it doesn't lend itself to the interpretation that `hrs_to_fund` = $1/\lambda$ = the average amount of time to fund a profile, which is used in $Exponential(\lambda)$. If not the mean, the median might also make sense.

So then, after 524 profiles we've learned about, our new $\lambda$ is the mean (or median?) of all the male hours to fund. Our new $\alpha$ is 1 + the number of male profiles we've learned about. So:

$$Gamma(1 + 524, 46 * \frac{1}{1 + 524} + \mu_{male} * (1 - \frac{1}{1 + 524}))$$

Wikipedia wants me to do:

$$Gamma(\alpha + n, \lambda + n\bar x)$$
$$Gamma(1 + 524, 1/46 + 524 \times \mu_{male})$$

The first formula I got myself, after one $Exp(\lambda)$, is:

$$Gamma(\alpha + 1, \lambda + x)$$

Starting from here, thinking intuitively about $n Exp(\lambda)$s, I get:

$$Gamma(\alpha + n, \lambda * \frac{\alpha}{\alpha + n} + \bar x * (1 - \frac{\alpha}{\alpha + n}))$$

```python
def plot_posterior_gamma(alpha, n, lambda_, xbar, a, b):
	alpha = float(alpha)
	plot_gamma(alpha + n, lambda_ * alpha / (alpha + n) + xbar * (1 - alpha / (alpha + n)), a=a, b=b)
```

```python
plot_posterior_gamma(alpha=1, 
                   n=df[df.gender == 'male'].shape[0], 
                   lambda_=df.hrs_to_fund.median(), 
                   xbar=df[df.gender == 'male'].hrs_to_fund.median(), 
                   a=10, b=40000)
```

##Using the Normal as a prior

Let's manually choose the values for our prior, what we believe to be the number of hours to fund a profile, using the Normal distibution.

```python
mean = df.hrs_to_fund.median()
variance = 50 # bigger is less sure, smaller is more sure, and smaller sigma

#sigma = np.sqrt(variance)
x = np.linspace(-200,200,100)
y = mlab.normpdf(x,mean,variance)
plt.plot(x,y)
plt.fill_between(x, 0, y, alpha=0.4)
plt.ylabel('Probability')
plt.xlabel('Hours to fund a profile')
```

Unlike with the Exponential distribution, we can state a more precise estimate for how many hours it takes to fund a profile. The main violation of the Normal distribution is that it allows negative hours to fund a profile, which is impossible. We'll compare the two results later on and see if the violation is major.

##Running the MCMC model

```python
lambda_ = 1./df.hrs_to_fund.median() # how many profiles funded per hour; rate of lambda_ profiles funded per hour; next profile funded in 1/lambda_ hours

# prior for mean male and mean female. i guess if i'm setting mu to the mean of both, that's empirical bayes

lambda_male = 1/df.hrs_to_fund[df.gender == 'male'].median()
lambda_female = 1/df.hrs_to_fund[df.gender == 'female'].median()

lambda_male = 1/df.hrs_to_fund[df.gender == 'male'].mean()
lambda_female = 1/df.hrs_to_fund[df.gender == 'female'].mean()

#p_male = pm.Normal('p_male', mu=80, tau=1./variance)
#p_female = pm.Normal('p_female', mu=80, tau=1./variance)

#mu = df.hrs_to_fund.mean()
#p_male = pm.Normal('p_male', mu=mu, tau=1./50)
#p_female = pm.Normal('p_female', mu=mu, tau=1./50)

p_male = pm.Gamma('p_male', 1, lambda_)
p_female = pm.Gamma('p_female', 1, lambda_)

@pm.deterministic
def delta(p_male=p_male, p_female=p_female):
    return p_male**-1 - p_female**-1

obs_male = pm.Exponential("obs_male", p_male, value=df.hrs_to_fund[df.gender == 'male'], observed=True)
obs_female = pm.Exponential("obs_female", p_female, value=df.hrs_to_fund[df.gender == 'female'], observed=True)

mcmc = pm.MCMC([p_male, p_female, delta, obs_male, obs_female])
mcmc.sample(20000, 1000)

p_male_samples = mcmc.trace("p_male")[:]**-1
p_female_samples = mcmc.trace("p_female")[:]**-1
delta_samples = mcmc.trace("delta")[:]
```

```python
with sns.plotting_context("paper", font_scale=2.5):
    plt.figure(figsize=(20,10))
    ax = plt.subplot(211)

    plt.hist(p_male_samples, histtype='stepfilled', bins=25, alpha=0.85,
             label="posterior of $1/\lambda_{male}$", color="#A60628", normed=True)
    plt.legend(loc="upper right")
    plt.title("Posterior distributions of $1/\lambda_{male}$, $1/\lambda_{female}$, and $1/\lambda_{male} - 1/\lambda_{female}$")
    
    plt.hist(p_female_samples, histtype='stepfilled', bins=25, alpha=0.85,
             label="posterior of $1/\lambda_{female}$", color="#467821", normed=True)
    plt.legend(loc="upper right")
    plt.xlabel('Average number of hours to fund a profile, $1/\lambda$')
    plt.tight_layout()

    ax = plt.subplot(212)
    plt.xlim(0, 40)
    plt.hist(delta_samples, histtype='stepfilled', bins=25, alpha=0.85,
             label="posterior of $1/\lambda_{male} - 1/\lambda_{female}$", color="#7A68A6", normed=True)
    plt.legend(loc="upper right")
    plt.xlabel('Difference between the average number of hours to fund a profile')
```
