#A Bayesian Approach: Does it takes longer to fund a male profile than a female profile?

A Bayesian approach allows us to make direct probability statements. The interpretation of its results is simpler than interpreting p-values or confidence intervals. 

The main drawback to a Bayesian approach is that we must specify our prior belief about the situation, and this inserts some subjectiveness into our calculation. Bayesians respond that we can create very weak priors that don't much affect the results, which will allow the data to dominate the final result.

In this case, our priors are a) the average number of hours to fund a male patient, and b) the average number of hours to fund a female patient. We will look at the data to inform our 'prior' belief; this method is called Empirical Bayes. We will also test various priors to see if they affect our final result much.

```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
sns.set_palette('Set2')
sns.set_context('paper', font_scale=2.0)
from matplotlib import pyplot as plt

from IPython.core.pylabtools import figsize
figsize(20,5)

import matplotlib.mlab as mlab
from scipy import stats
import pymc as pm
from sklearn.linear_model import LinearRegression
import matplotlib.patches as mpatches
```

```python
df = pd.read_csv('data.csv')
df['target_amt'] = df.target_amount * .01
```

The data we have is the amount of time it takes to fund a profile. For each profile, we can think of the time it takes to be funded as an Exponential random variable. The story for the Exponential, as taken from [the notes for](https://bayesrule.files.wordpress.com/2014/07/probability_cheatsheet_140718.pdf) Joe Blitzstein's [Intro to Probability course](http://www.stat110.net), is:

> You are in a grassy field at night, looking at shooting stars. The time in between each shooting star can be modeled as $Exponential(\beta)$, where $\beta$ is the expected amount of time until a shooting star passes.

We can think of our data, the number of hours to fund a profile, as coming out of a random number generator. This number generator is given the expected number of hours for a profile to be funded, and it gives us sample numbers. In this sense, we view the data as a random variable. (In the classical context, the data is not random. It either is a certain figure, or it isn't.)

##Assumption violations of the Exponential distribution

In the example, we are measuring the time between each shooting star. But when we measure time-to-fund, it's not that one profile is posted onto the website, we measure how long it takes to be funded, then after it's funded, another profile is posted, and we measure the time for that to be funded. Instead, there are multiple profiles posted to the site, and we are starting our timer from when the profile is posted to when it is funded. In the lingo, we are not strictly looking at inter-arrival times.

An important property of the Exponential distribution is that it is memoryless. The time it took for one shooting star doesn't affect the time until the next shooting star. But for us, a donor might fully fund multiple profiles in a session, causing the latter profiles' time-to-fund to be affected by its predecessors'. This violation of the memoryless property stems from the fact that we're not strictly measuring interarrival times.

Another assumption we violate is that the times-to-fund are identically distributed. To us sitting in a grassy field, all the shooting stars are the same. A certain kind of shooting star does not occur more frequently than another. But our profiles can more easily be categorized into various groups. For example, profiles that request more money take longer to fund.

Even though our data violates the assumptions of the Exponential distribution, and more broadly, the Poisson process, it is close enough that the results we get are applicable to our problem.

#Method I: Using conjugate priors, Gamma distribution

If our data fits an Exponential distribution, what distribution should we choose to state our prior belief of the number of hours to fund a male profile and a female profile? If we use a conjugate prior, then our final result will be analytically tractable, and we will be able to come to a mathematically clean solution.

The conjugate prior here is the $Gamma(\alpha,\beta)$ distribution, where $\beta$ is the same as in the $Exponential(\beta)$ distribution.

The inutitive story for the Gamma distribution is:

> We are again looking at shooting stars. We want to wait to see $\alpha$ stars before we go home. The amount of time between each shooting star can be modeled as $Exponential(\beta)$, where $\beta$ is the expected amount of time until a shooting star passes. The probability that all $\alpha$ shooting stars will occur by $X$ total time is modeled by $Gamma(\alpha, \beta)$.

Let's model the time to fund a single profile via the $Gamma(\alpha, \beta)$ distribution, where $\alpha=1$. If we want to see how long it takes for one profile to be funded, that is the same story as for the $Exponential(\beta)$ distribution. $Gamma(1, \beta) = Exponential(\beta)$.

```python
def plot_gamma(num_profiles, hrs_to_fund, a=0, b=100):
	"""num_profiles is # of occurrences. hrs_to_fund is E[X] for Expo(lambda). a and b are x-axis limits.
    hrs_to_fund = 1/lambda_!!!"""
	x = np.linspace(a, b, 10000)
	y = stats.gamma.pdf(x, a=num_profiles, scale=hrs_to_fund)

	plt.plot(x, y)
	plt.fill_between(x, 0, y, alpha=0.4)
	plt.xlabel('hours until all {} profiles funded when {:.2f} hours until the next profile funded'.format(num_profiles, hrs_to_fund))
	plt.ylabel('Probability')
    
	print stats.gamma.cdf(1, a=num_profiles, scale=hrs_to_fund), 'probability after 1 hr'
	print stats.gamma.cdf(num_profiles*hrs_to_fund, a=num_profiles, scale=hrs_to_fund), 'probability after {} hrs'.format(hrs_to_fund*num_profiles)
	print 'MAP is at X = {}; MAP/a = {}'.format(x[y.argmax()], x[y.argmax()] / num_profiles)
```

```python
plot_gamma(num_profiles=1, hrs_to_fund=df.hrs_to_fund.mean())
```

```python
hrs_to_fund = df.hrs_to_fund.mean()

x = np.linspace(0, 100, 10000)
y = stats.expon.pdf(x, scale=hrs_to_fund)

plt.plot(x, y)
plt.fill_between(x, 0, y, alpha=0.4)
plt.xlabel('hours until the next occurence when we expect {:.2f} hours until the next occurrence'.format(hrs_to_fund))
plt.ylabel('Probability')

print stats.expon.cdf(1, scale=hrs_to_fund), 'probability after 1 hr'
print stats.expon.cdf(hrs_to_fund, scale=hrs_to_fund), 'probability after {} hr'.format(hrs_to_fund)
```

#Choosing our prior parameter values via Empirical Bayes

For both our prior belief for the number of hours to fund a male profile, and the number of hours to fund a female profile, we will say that we know about 1 profile, and we will set the expected time for that that profile to be funded to be the median (or mean?) of all the profiles. We will choose:

$$Gamma(\alpha = 1, \beta = 46.36)$$

The fact that we are looking at our post-experiment data to choose our 'prior' belief as to the number of hours to fund a profile is an example of the Empirical Bayes technique. Traditionally, we should not use post-experiment data to choose our priors.

```python
plot_gamma(1, df.hrs_to_fund.median())
```

I chose the median instead of the mean for the expected value of how many hours until a profile is funded because I think it's less affected by extreme values and the high variance that this distribution has.

I chose 1 profile arbitrarily. If $\alpha=1$, the prior is not very strong, which is what I want. I could also give more weight to my prior, by increasing the number of profile I've 'seen' before the experiement. This is where subjective priors come into play.

I also started with a prior assumption that male and female profiles take the same amount of time to fund. I could have made one bigger than the other.

Let's check later how much of a difference it makes to give more weight to my prior, and/or to make our priors different for the male and female groups.

#Our updated beliefs after we see the data, the posterior distribution

Our prior belief was that it would take 46 hours to fund a profile, for both males and females, having 'seen' 1 profile. The data we've collected shows us how long it took to fund each male profile, and each female profile. Now, we can combine our prior belief with the data to get the probability that it would take $X$ hours to fund all our 524 male profiles, and similarly for all our 584 female profiles. This probability distribution is called the posterior distribution.

Our posterior distribution here, because I used the $Gamma(\alpha, \beta)$ conjugate prior, is $Gamma(\alpha + 1, \frac{\beta + x}{2})$ for one $Exponential(\beta)$ experiment when $\alpha = 1$ and $n = 1$, or we started with one new profile that I've learned about and we've learned about one additional profile.

So, let's say that we have a new $Exponential(\beta)$ profile, and it took 75 hours to fund it.

My updated posterior distribution is $Gamma(1 + 1, \frac{46 + 75}{2})$.

```python
plot_gamma(2, (46 + 75)/2., b=100)
```

###Statements we can make

Before our result of 75 hours, our expected value for the number of hours to fund a profile was 46 hours. Now, we expect it to take 60.5 hours to fund a profile.

```python
stats.gamma.cdf(48, a=1, scale=46)
```

```python
stats.gamma.cdf(48, a=2, scale=(46 + 75)/2.) * 2
```

With our prior, there was a 65% chance that 1 profile would be funded within 2 days. Now, there's an 38% chance that 1 profile is funded within 2 days.

###More general form of posterior distribution

So then, after 524 profiles we've learned about, our new $\lambda$ is the mean (or median?) of all the male hours to fund, along with our prior. Our new $\alpha$ is 1 + the number of male profiles we've learned about. So:

$$Gamma(1 + 524, 46 * \frac{1}{1 + 524} + \bar x_{male} * (1 - \frac{1}{1 + 524}))$$

It seems to make more sense to have the new `hrs_to_fund` be an average of the two figures, rather than just a sum. A sum will blow up when `n` gets large, and it doesn't lend itself to the interpretation that `hrs_to_fund` = $1/\lambda$ = the average amount of time to fund a profile, which is used in $Exponential(\lambda)$. If not the mean, the median might also make sense.

The first formula I got myself, after one $Exp(\beta)$, is:

$$Gamma(\alpha + 1, \beta + x)$$

But it makes more sense intuitively as:

$$Gamma(\alpha + 1, \frac{\beta + x}{2})$$

Wikipedia wants me to do:

$$Gamma(\alpha + n, \beta + n\bar x)$$
$$Gamma(1 + 524, 1/46 + 524 \times \bar x_{male})$$

Thinking intuitively about $n\ Exp(\beta)$s, I get:

$$Gamma(\alpha + n, \beta * \frac{\alpha}{\alpha + n} + \bar x * (1 - \frac{\alpha}{\alpha + n}))$$

where $x$ is the number of hours to fund a profile.

###Results after inputting our data

```python
def plot_posterior_gamma(alpha, n, lambda_, xbar, a, b):
	alpha = float(alpha)
	plot_gamma(alpha + n, lambda_ * alpha / (alpha + n) + xbar * (1 - alpha / (alpha + n)), a=a, b=b)
```

```python
plot_posterior_gamma(alpha = 1, 
                   n = df[df.gender == 'male'].shape[0], 
                   lambda_ = df.hrs_to_fund.mean(), 
                   xbar = df[df.gender == 'male'].hrs_to_fund.mean(), 
                   a=10, b=60000)
```

This is our probability distribution for the number of hours it takes to fund all the 525 male profiles. The most likely amount of time it takes to fund all the profiles is called the Maximum a Posteriori probability, the number of hours with the highest chance of happening.

The most likely amount of time to fund all male profiles is 48,331 hours. If there are 525 profiles in total, the most likely amount of time to fund one profile is 92 hours.

####And for female profiles...

```python
plot_posterior_gamma(alpha = 1, 
                   n = df[df.gender == 'female'].shape[0], 
                   lambda_ = df.hrs_to_fund.median(), 
                   xbar = df[df.gender == 'female'].hrs_to_fund.median(), 
                   a=10, b=60000)
```

The most likely amount of time to fund all female profiles is 24,158 hours. If there are 584 profiles in total, the most likely amount of time to fund one profile is 41 hours.

And now, both the posterior probability distibutions plotted together:

```python
def plot_multiple_posterior_gamma(label, color, alpha, n, lambda_, xbar, a, b):
	alpha = float(alpha)

	def plot_gamma(num_profiles, hrs_to_fund, a=0, b=100):
		"""num_profiles is # of occurrences. hrs_to_fund is E[X] for Expo(lambda). a and b are x-axis limits.
	    hrs_to_fund = 1/lambda_!!!"""
		x = np.linspace(a, b, 10000)
		lambda_ = 1./hrs_to_fund # how many profiles funded per hour; rate of lambda_ profiles funded per hour; next profile funded in 1/lambda_ hours
		y = stats.gamma.pdf(x, a=num_profiles, scale=1./lambda_)

		plt.plot(x, y, label=label, color=color)
		plt.fill_between(x, 0, y, alpha=0.4, color=color)
		plt.xlabel('hours until all profiles funded')
		plt.ylabel('Probability')
		plt.legend(loc='upper right')

	plot_gamma(alpha + n, lambda_ * alpha / (alpha + n) + xbar * (1 - alpha / (alpha + n)), a=a, b=b)
	labels.append(mpatches.Patch(color=color, label=label))
	plt.legend(handles=labels)
```

```python
labels = []

plot_multiple_posterior_gamma(label='Male',
    color = sns.color_palette()[0],
    alpha = 1, 
    n = df[df.gender == 'male'].shape[0], 
    lambda_ = df.hrs_to_fund.mean(), 
    xbar = df[df.gender == 'male'].hrs_to_fund.mean(), 
    a=35000, b=55000)

plot_multiple_posterior_gamma(label='Female',
    color = sns.color_palette()[2],
    alpha = 1, 
    n = df[df.gender == 'female'].shape[0], 
    lambda_ = df.hrs_to_fund.mean(), 
    xbar = df[df.gender == 'female'].hrs_to_fund.mean(), 
    a=35000, b=55000)
```

If we have the probability distribution of the hours until all the profiles are funded, we can divide each distribution by the number of profiles each to get the probability of the number of hours until 1 profile is funded:

```python
def plot_multiple_posterior_gamma(label, color, alpha, n, lambda_, xbar, a, b):
	alpha = float(alpha)

	def plot_gamma(num_profiles, hrs_to_fund, a=0, b=100):
		"""num_profiles is # of occurrences. hrs_to_fund is E[X] for Expo(lambda). a and b are x-axis limits.
	    hrs_to_fund = 1/lambda_!!!"""
		x = np.linspace(a, b, 10000)
		lambda_ = 1./hrs_to_fund # how many profiles funded per hour; rate of lambda_ profiles funded per hour; next profile funded in 1/lambda_ hours
		y = stats.gamma.pdf(x, a=num_profiles, scale=1./lambda_)

		plt.plot(x/n, y, label=label, color=color)
		plt.fill_between(x/n, 0, y, alpha=0.4, color=color)
		plt.xlabel('hours until 1 profile funded')
		plt.ylabel('Probability')
		plt.legend(loc='upper right')

	plot_gamma(alpha + n, lambda_ * alpha / (alpha + n) + xbar * (1 - alpha / (alpha + n)), a=a, b=b)
	labels.append(mpatches.Patch(color=color, label=label))
	plt.legend(handles=labels)

labels = []

plot_multiple_posterior_gamma(label='Male',
    color = sns.color_palette()[0],
    alpha = 1, 
    n = df[df.gender == 'male'].shape[0], 
    lambda_ = df.hrs_to_fund.mean(), 
    xbar = df[df.gender == 'male'].hrs_to_fund.mean(), 
    a=35000, b=55000)

plot_multiple_posterior_gamma(label='Female',
    color = sns.color_palette()[2],
    alpha = 1, 
    n = df[df.gender == 'female'].shape[0], 
    lambda_ = df.hrs_to_fund.mean(), 
    xbar = df[df.gender == 'female'].hrs_to_fund.mean(), 
    a=35000, b=55000)
```

###Interpretation

Although this graph looks identical to our classical estimate for true means via sample means, it is interpreted differently. Here, we can make direct probability statements, without thinking about long-run frequency. Our results here pertain to the data we just collected, not to hypothetical future experiments and the mean of those. 

#Method II: Using non-conjugate priors via MCMC, the Normal

Let's manually choose the values for our prior, what we believe to be the number of hours to fund a profile, using the Normal distibution.

```python
mean = df.hrs_to_fund.median()
variance = 50 # bigger is less sure, smaller is more sure, and smaller sigma

#sigma = np.sqrt(variance)
x = np.linspace(-200,200,100)
y = mlab.normpdf(x,mean,variance)
plt.plot(x,y)
plt.fill_between(x, 0, y, alpha=0.4)
plt.ylabel('Probability')
plt.xlabel('Hours to fund a profile')
```

Unlike with the Exponential distribution, we can state a more precise estimate for how many hours it takes to fund a profile. The main violation of the Normal distribution is that it allows negative hours to fund a profile, which is impossible. We'll compare the two results later on and see if the violation is major.

##Running the MCMC model

```python
lambda_ = 1./df.hrs_to_fund.median() # how many profiles funded per hour; rate of lambda_ profiles funded per hour; next profile funded in 1/lambda_ hours

# prior for mean male and mean female. i guess if i'm setting mu to the mean of both, that's empirical bayes

lambda_male = 1/df.hrs_to_fund[df.gender == 'male'].median()
lambda_female = 1/df.hrs_to_fund[df.gender == 'female'].median()

lambda_male = 1/df.hrs_to_fund[df.gender == 'male'].mean()
lambda_female = 1/df.hrs_to_fund[df.gender == 'female'].mean()

#p_male = pm.Normal('p_male', mu=80, tau=1./variance)
#p_female = pm.Normal('p_female', mu=80, tau=1./variance)

#mu = df.hrs_to_fund.mean()
#p_male = pm.Normal('p_male', mu=mu, tau=1./50)
#p_female = pm.Normal('p_female', mu=mu, tau=1./50)

p_male = pm.Gamma('p_male', 1, lambda_)
p_female = pm.Gamma('p_female', 1, lambda_)

@pm.deterministic
def delta(p_male=p_male, p_female=p_female):
    return p_male**-1 - p_female**-1

obs_male = pm.Exponential("obs_male", p_male, value=df.hrs_to_fund[df.gender == 'male'], observed=True)
obs_female = pm.Exponential("obs_female", p_female, value=df.hrs_to_fund[df.gender == 'female'], observed=True)

mcmc = pm.MCMC([p_male, p_female, delta, obs_male, obs_female])
mcmc.sample(20000, 1000)

p_male_samples = mcmc.trace("p_male")[:]**-1
p_female_samples = mcmc.trace("p_female")[:]**-1
delta_samples = mcmc.trace("delta")[:]
```

```python
with sns.plotting_context("paper", font_scale=2.5):
    plt.figure(figsize=(20,10))
    ax = plt.subplot(211)

    plt.hist(p_male_samples, histtype='stepfilled', bins=100, alpha=0.85,
             label="posterior of $\beta_{male}$", color=sns.color_palette()[0], normed=True)
    plt.legend(loc="upper right")
    plt.title("Posterior distributions of $\beta_{male}$, $\beta_{female}$, and $\beta_{male} - \beta_{female}$")
    
    plt.hist(p_female_samples, histtype='stepfilled', bins=100, alpha=0.85,
             label="posterior of $\beta_{female}$", color=sns.color_palette()[2], normed=True)
    plt.legend(loc="upper right")
    plt.xlabel('Average number of hours to fund a profile, $\beta$')
    plt.tight_layout()

    ax = plt.subplot(212)
    plt.xlim(0, 40)
    plt.hist(delta_samples, histtype='stepfilled', bins=100, alpha=0.85,
             label="posterior of $\beta_{male} - \beta_{female}$", color=sns.color_palette()[3], normed=True)
    plt.legend(loc="upper right")
    plt.xlabel('Difference between the average number of hours to fund a profile')
```
