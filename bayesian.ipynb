#A Bayesian Approach: What is the probability that a male patient takes longer to fund than a female patient?

A Bayesian approach allows us to make direct probability statements, such as the one in the title. The interpretation of its results is simpler than interpreting p-values or confidence intervals. 

The main drawback to a Bayesian approach is that we must specify our prior belief about our unknown parameter, and that this inserts some subjectiveness into our scientific process. Bayesians respond that we can create very weak priors that don't much affect the results, priors that will allow the data to dominate the final result.

In this case, our priors are a) the number of hours to fund an average male patient, and b) the number of hours to fund an average female patient. We will test various priors to see if they affect our final result much

```python
import pymc as pm
import pandas as pd

# what's my prior on the diff btwn time to fund male and female?
#p = pm.Normal('p', mu=0, tau=1./variance)
# this one is actually not used

# prior for mean male and mean female. i guess if i'm setting mu to the mean of both, that's empirical bayes
#p_male = pm.Normal('p_male', mu=80, tau=1./variance)
#p_female = pm.Normal('p_female', mu=80, tau=1./variance)

p_male = pm.Exponential('p_male', lambda_)
p_female = pm.Exponential('p_female', lambda_)

#p_male = pm.Gamma('p_male', df.shape[0], lambda_)
#p_female = pm.Gamma('p_female', df.shape[0], lambda_)

@pm.deterministic
def delta(p_male=p_male, p_female=p_female):
    return p_male - p_female

df = pd.read_csv('data.csv')

# so the prior was a Normal rv estimating the # of hrs to fund a profile.
# our observed data is a Poisson rv estimating,
# like the # of emails in a day,
# the number of hrs to fund a profile.
# so there is an expected # of hrs to fund a profile.

# not exactly sure if the data qualifies as Poisson
obs_male = pm.Exponential("obs_male", p_male, value=1./df.hrs_to_fund[df.gender == 'male'], observed=True)
obs_female = pm.Exponential("obs_female", p_female, value=1./df.hrs_to_fund[df.gender == 'female'][df.target_amount != 21500], observed=True)

mcmc = pm.MCMC([p_male, p_female, delta, obs_male, obs_female])
mcmc.sample(20000, 1000)

p_male_samples = mcmc.trace("p_male")[:]
p_female_samples = mcmc.trace("p_female")[:]
delta_samples = mcmc.trace("delta")[:]
```

```python
with sns.axes_style("darkgrid"):
    plt.figure(figsize=(20,10))
    ax = plt.subplot(311)

    #plt.xlim(90, 95)
    plt.hist(p_male_samples, histtype='stepfilled', bins=25, alpha=0.85,
             label="posterior of $p_{male}$", color="#A60628", normed=True)
    plt.legend(loc="upper right")
    plt.title("Posterior distributions of $p_{male}$, $p_{female}$, and $p_{male} - p_{female}$")

    ax = plt.subplot(312)

    #plt.xlim(70, 75)
    plt.hist(p_female_samples, histtype='stepfilled', bins=25, alpha=0.85,
             label="posterior of $p_{female}$", color="#467821", normed=True)
    plt.legend(loc="upper right")

    ax = plt.subplot(313)
    #plt.xlim(17, 23)
    plt.hist(delta_samples, histtype='stepfilled', bins=30, alpha=0.85,
             label="posterior of $p_{male} - p_{female}$", color="#7A68A6", normed=True)
    plt.legend(loc="upper right");
```
